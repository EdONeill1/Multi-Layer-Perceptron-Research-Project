# XOR Experiments 8-15

```python
import numpy as np
import random 
import os
import json
import mlp


np.random.seed(2021)
random.seed(2021)

path = ""
save_logs = True
print_graphs = True
task = "XOR"
verbose = True
return_prediction = True


def xor(x,y):
    return x ^ y

X = np.array([[0,0], [0,1], [1,0], [1,1]])
Y = np.array([[0,1,1,0]]).T
```

## Experiment 8
* Learning rate = 1
* Epochs trained for = 1000000
* Hidden Units = 4

```python
np.random.seed(2021)
random.seed(2021)

lr = 1
epochs = 1000000
hidden_units = 4
experiment = 8
task = "XOR"

W1, W2 = mlp.init_weights(X, Y, hidden_units)


errors8 = []
error_to_save = []
acc_to_save = []
acc8 = []

accuracy = 0
for epoch in range(epochs):

    A1, A2, error = mlp.forward_prop(W1, W2, X, Y, mlp.sigmoid)
    dW1, dW2 = mlp.back_prop(X, Y, A1, A2, W1, W2, mlp.sigmoid_derivative)
    W1, W2 = mlp.update_weights(W1, W2, dW1, dW2, lr)

    accuracy = 100 * (1 - np.mean(np.abs(Y - np.where(A2 < 0.5, 0, 1))))

    if accuracy < 0:
        accuracy = 0
    errors8.append(error)
    acc8.append(accuracy)
   

    if epoch % (epochs/10) == 0:
                     error_to_save.append(error)
                     acc_to_save.append(accuracy)
                     mlp.print_status(epoch, epochs, error, accuracy)
                     mlp.save_logs(path, save_logs, task, error, accuracy, int(((epoch/epochs) * 100) + 10), 
                               epochs, hidden_units, lr, experiment)

if print_graphs == True:
    mlp.plot_result(task, "Error", errors8, acc8, lr, hidden_units, epochs, experiment)
    mlp.plot_result(task, "Accuracy", errors8, acc8, lr, hidden_units, epochs, experiment)


    
A2, prediction_error, prediction_accuracy = mlp.predict(task, X, Y, W1, W2, mlp.sigmoid, verbose, True)
mlp.save_prediction(path, save_logs, task, prediction_error, prediction_accuracy, epochs, hidden_units, lr, experiment)
```
