# XOR Experiments With Large Input and Output - Experiments 16-18

```python
import numpy as np
import random 
import os
import json
import mlp

np.random.seed(2021)
random.seed(2021)

path = ""
save_logs = True
print_graphs = True
task = "XOR"
verbose = True
return_prediction = True


N = 500

def xor(x,y):
    return x ^ y

X = np.array([[0,0], [0,1], [1,0], [1,1]])
Y = np.array([[0,1,1,0]]).T
```

## Retraining MLP from Experiment 6

```python
np.random.seed(2021)
random.seed(2021)

lr = 1
epochs = 1000000
hidden_units = 4
experiment = None

W1, W2 = mlp.init_weights(X, Y, hidden_units)

mlp.train(task, X, Y, W1, W2, mlp.sigmoid, mlp.sigmoid_derivative, 
          epochs, lr, hidden_units, experiment, path, save_logs, print_graphs)

_, prediction_error, prediction_accuracy = mlp.predict(task, X, Y, W1, W2, mlp.sigmoid, verbose, return_prediction)
mlp.save_prediction(path, save_logs, task, prediction_error, prediction_accuracy, epochs, hidden_units, lr, experiment)
```

### The Large Input Output Data

```python
X = np.array([[random.randrange(0,2), random.randrange(0,2)] for i in range(N)])

Y = np.array([[xor(X[i][0], X[i][1]) for i in range(len(X))]]).T

X_train, X_test, Y_train, Y_test = (X[:int(0.8 * len(X))],
                                    X[int(0.8 * len(X)):],
                                    Y[:int(0.8 * len(Y))], 
                                    Y[int(0.8 * len(Y)):])
```

### Using pretrained weights from experiment six on the large X,Y

```python
mlp.predict(task, X, Y, W1, W2, mlp.sigmoid, verbose, return_prediction=False)
```

## Experiment 16 - Training Large Input With It's Own Weights
* Learning rate = 1
* Epochs trained for = 1000000
* Hidden Units = 4

```python
np.random.seed(2021)
random.seed(2021)

lr = 1
epochs = 1000000
hidden_units = 4
experiment = 16
task = "XOR"

W1, W2 = mlp.init_weights(X, Y, hidden_units)

mlp.train(task, X_train, Y_train, W1, W2, mlp.sigmoid, mlp.sigmoid_derivative, 
          epochs, lr, hidden_units, experiment, path, save_logs, print_graphs)

_, prediction_error, prediction_accuracy = mlp.predict(task, X_test, Y_test, W1, W2, mlp.sigmoid, verbose, return_prediction)
mlp.save_prediction(path, save_logs, task, prediction_error, prediction_accuracy, epochs, hidden_units, lr, experiment)
```

## Experiment 17 - Reducing Learning Rate To Try For Better Results
* Learning rate = 0.001
* Epochs trained for = 1000000
* Hidden Units = 4

```python
np.random.seed(2021)
random.seed(2021)

lr = 0.001
epochs = 1000000
hidden_units = 4
experiment = 17
task = "XOR"

W1, W2 = mlp.init_weights(X, Y, hidden_units)

mlp.train(task, X_train, Y_train, W1, W2, mlp.sigmoid, mlp.sigmoid_derivative, 
          epochs, lr, hidden_units, experiment, path, save_logs, print_graphs)

_, prediction_error, prediction_accuracy = mlp.predict(task, X_test, Y_test, W1, W2, mlp.sigmoid, verbose, return_prediction)
mlp.save_prediction(path, save_logs, task, prediction_error, prediction_accuracy, epochs, hidden_units, lr, experiment)
```
