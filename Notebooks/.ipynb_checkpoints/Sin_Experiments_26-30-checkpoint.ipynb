# Sin Experiments 26-30


```python
import numpy as np
import random 
import os
import json
import mlp


np.random.seed(2021)
random.seed(2021)

path = ""

save_logs = True
print_graphs = True
task = "Sin"
verbose = False
return_prediction = True


def sin(x):
    y = x[0] - x[1] + x[2] - x[3]
    return np.sin(y)
```

### Creating Input And Output Data

```python
N = 500

X = np.array([[random.randrange(-1,2), random.randrange(-1,2),
 random.randrange(-1,2), random.randrange(-1,2)] for i in range(N)])
Y = np.array([[sin(i) for i in X]]).T

X_train, X_test, Y_train, Y_test = (X[:int(0.8 * len(X))],
                                    X[int(0.8 * len(X)):],
                                    Y[:int(0.8 * len(Y))], 
                                    Y[int(0.8 * len(Y)):])
```

## Experiment 26
* Learning rate = 0.001
* Epochs trained for = 1000000
* Hidden Units = 6

```python
np.random.seed(2021)
random.seed(2021)

lr = 0.001
epochs = 1000000
experiment = 26
hidden_units = 6

W1, W2 = mlp.init_weights(X, Y, hidden_units)

mlp.train(task, X_train, Y_train, W1, W2, mlp.tanh, mlp.tanh_derivative, 
          epochs, lr, hidden_units, experiment, path, save_logs, print_graphs)
```

```python
A2, prediction_error, prediction_accuracy = mlp.predict(task, X_test, Y_test, W1, W2, mlp.tanh, verbose, return_prediction)
mlp.save_prediction(path, save_logs, task, prediction_error, prediction_accuracy, epochs, hidden_units, lr, experiment)
mlp.compare_prediction(A2, Y_test, "Sin", experiment)
```
