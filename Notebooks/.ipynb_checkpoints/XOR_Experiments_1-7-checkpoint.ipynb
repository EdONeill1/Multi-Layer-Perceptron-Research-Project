# XOR Experiments 1-7

```python
import numpy as np
import random 
import os
import json
import mlp


np.random.seed(2021)
random.seed(2021)

path = ""
save_logs = True
print_graphs = True
task = "XOR"
verbose = True
return_prediction = True


def xor(x,y):
    return x ^ y


X = np.array([[0,0], [0,1], [1,0], [1,1]])
Y = np.array([[0,1,1,0]]).T
```

## Experiment 1
* Learning rate = 1
* Epochs trained for = 100
* Hidden Units = 3

```python
np.random.seed(2021)
random.seed(2021)

lr = 1
epochs = 100
hidden_units = 3
experiment = 1

W1, W2 = mlp.init_weights(X, Y, hidden_units)

mlp.train(task, X, Y, W1, W2, mlp.sigmoid, mlp.sigmoid_derivative, 
          epochs, lr, hidden_units, experiment, path, save_logs, print_graphs)

_, prediction_error, prediction_accuracy = mlp.predict(task, X, Y, W1, W2, mlp.sigmoid, verbose, return_prediction)
mlp.save_prediction(path, save_logs, task, prediction_error, prediction_accuracy, epochs, hidden_units, lr, experiment)
```
